```python
import pandas as pd

# CSV 파일 읽어오기
df = pd.read_csv("address_with_all_codes.csv", dtype=str)

# 문제 있는 행을 담을 리스트
bad_rows = []

for idx, row in df.iterrows():
    tokens = row["검색어"].split()            # ex. ["서울특별시","종로구"]
    link_text = row["link_text"]             # ex. "서울특별시, 종로구, 부암동"
    # 토큰 중 link_text에 없는 것들
    missing = [t for t in tokens if t not in link_text]
    if missing:
        bad_rows.append((idx, row.to_dict(), missing))

# 결과 출력
if not bad_rows:
    print("모든 행이 정상입니다.")
else:
    for idx, rowdict, missing in bad_rows:
        print(f"★ 행 {idx} 오류:")
        print(f"  검색어 tokens : {rowdict['검색어'].split()}")
        print(f"  link_text    : {rowdict['link_text']}")
        print(f"  누락된 토큰  : {missing}")
        print(f"  전체 데이터  : {rowdict}")
        print("-" * 40)
```
